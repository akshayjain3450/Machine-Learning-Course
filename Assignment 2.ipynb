# Level Surface Plot and 3-D Surface Plot for X^2 + Y^2

Libraries

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Generate Paramters and Vectorisation

```python
w0 = np.linspace(-10, 10, 100)
w1 = np.linspace(-10, 10, 100)
W0, W1 = np.meshgrid(w0,w1)
```

Equation : X^2 + Y^2

Calculation function value

```python
func = W0**2 + W1**2
#print(func)
```

Level Surface Plot

```python
plt.contour(W0,W1,func)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph showing the Equation: W0 ^ 2   + W1 ^ 2")
plt.grid(True)
```

3-D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(W0, W1, func,cmap='viridis', edgecolor='none')
ax.set_title('3D-Surface plot')
plt.show()
```

# Level Surface Plot and 3-D Surface Plot for Error Function

Libraries

```python
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Data Set

```python
m = 2 # Slope is 2
c = 2 # Intercept is 2
x = np.linspace(1, 10, 11)
y = m * x + c
#print(x)
#print(y)
```

Plotting Line Y = m*X + C with intercept as 2

```python
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title(" Linear Model y = mx + c where c = 2 ")
plt.scatter(x,y)
plt.ylim(bottom = 0, top = 24)
plt.show()
```

Generate Parameters

```python
def generate_parameterw0():
    w0 = np.linspace(-10, 10, 100)
    return w0
```

```python
def generate_parameterw1():
    w1 = np.linspace(-10, 10, 100)
    return w1
```

Vectorisation

```python
W0, W1 = np.meshgrid(generate_parameterw0(),generate_parameterw1())
```

Error Function

```python
def error_func():
    e = 0
    for i in range(len(x)):
        e += (W1 * x[i] + W0 - y[i]) ** 2
    return e
```

Level Surface Plot

```python
lvl = np.linspace(-100,100,20)
plt.xlabel("W0")
plt.ylabel("W1")
plt.title("Error Function on two parameters W0(intercept) and W1(slope) \n")
plt.contour(W0,W1,error_func(),levels = lvl )
plt.ylim(bottom = 0, top = 5)
plt.xlim(left = -4, right = 7)
plt.grid(True)
```

3-D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(W0, W1, error_func(),cmap='autumn', edgecolor='none')
ax.set_title('3D-Surface plot between parameters slope (W1) and intercept (W0) and error function')
plt.show()
```
